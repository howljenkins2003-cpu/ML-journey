```python
import pandas as pd

df = pd.read_csv("HR.csv")
print(df.shape)
print(df.head())
```

```python
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

```python
print(df.columns.tolist())
```

```python
suspicious = ['Education', 'JobLevel', 'EnvironmentSatisfaction', 
              'JobSatisfaction', 'WorkLifeBalance', 'PerformanceRating',
              'RelationshipSatisfaction', 'JobInvolvement', 'StockOptionLevel']

for col in suspicious:
    print(f"{col}: {sorted(df[col].unique())}")
```

```python
print(df["Attrition"].value_counts())
```

```python
print(df[["EmployeeCount", "Over18", "StandardHours"]].nunique())
```

```python
df = df.drop(columns = ["EmployeeCount", "Over18", "StandardHours"])
print(df.shape)
```

```python
df = df.drop(columns= ["EmployeeNumber"])
```

```python
str_cols = df.select_dtypes(include='str').columns.tolist()
print(str_cols)
for col in str_cols:
    print(f"{col}: {df[col].nunique()} — {df[col].unique()}")
```

```python
df['IncomePerLevel'] = df['MonthlyIncome'] / df['JobLevel']
print(df[['MonthlyIncome', 'JobLevel', 'IncomePerLevel']].head(10))
```

```python
df['PromotionStagnation'] = df['YearsSinceLastPromotion'] / (df['YearsAtCompany'] + 1)
```

```python
print(df[['YearsSinceLastPromotion', 'YearsAtCompany', 'PromotionStagnation']].head(10))
```

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
print(df['Attrition'].value_counts())
```
